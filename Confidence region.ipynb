```python
import pandas as pd
import numpy as np 
import pickle as pickle
import gudhi as gd
import seaborn as sns
from scipy.spatial import distance_matrix
from pylab import *
```

```python
U1 = np.random.uniform(0, 2 * pi, size= 1000)
V1 = np.array([[0.35 * cos(u) + 0.02 * np.random.uniform(-1,1), 0.35 * sin(u) + 0.02 * np.random.uniform(-1,1)] for u in U1])

U2 = np.random.uniform(0, 2 * pi, size= 2000)
V2 = np.array([[0.7 * cos(u) + 0.02 * np.random.uniform(-1,1), 0.7 * sin(u) + 0.02 * np.random.uniform(-1,1)] for u in U2])

W = np.concatenate((V1, V2), axis=0)
```

```python
plt.scatter(W[:,0], W[:,1], s = 0.1)
plt.show()
```

```python
alpha_complex_W = gd.AlphaComplex(points = W)
simplex_tree_W = alpha_complex_W.create_simplex_tree()
```

```python
filtration_list_W = simplex_tree_W.get_filtration()
```

```python
for simplex in filtration_list_W:
    simplex_tree_W.assign_filtration(simplex[0], filtration = np.sqrt(simplex[1]))
```

```python
persistence_W = simplex_tree_W.persistence()
```

```python
gd.plot_persistence_diagram(persistence_W)
plt.show()
```

```python
gd.plot_persistence_diagram(persistence_W, band = 2 *0.05)
plt.show()
```
